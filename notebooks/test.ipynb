```python
import os, sys, findspark
from pyspark.sql import SparkSession
```

```python
os.environ["JAVA_HOME"] = "/opt/homebrew/Cellar/openjdk@17/17.0.17/libexec/openjdk.jdk/Contents/Home"
os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable
findspark.init()
spark = (
    SparkSession.builder
    .appName("SpotifyRec")
    .master("local[4]")
    .config("spark.driver.memory", "10g")
    .config("spark.sql.adaptive.enabled", "true")
    .getOrCreate()
)
```

```python
pt = spark.read.parquet("../parquet_data/playlist_tracks")
pt.createOrReplaceTempView("playlist_tracks")

p = spark.read.parquet("../parquet_data/playlists")
p.createOrReplaceTempView("playlists")

t = spark.read.parquet("../parquet_data/tracks")
t.createOrReplaceTempView("tracks")
```

```python
spark.sql("""
SELECT *
FROM playlist_tracks
LIMIT 10
""").show(truncate=False)
```
